```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add nodes with labels and categories
G.add_node("Prompting Techniques", size=3000)
G.add_node("Basic Prompting Methods", size=2000)
G.add_node("Advanced Prompting Techniques", size=2000)
G.add_node("Interactive & Reasoning-Driven Methods", size=2000)

G.add_node("One-Shot Prompting", size=1000)
G.add_node("Few-Shot Prompting", size=1000)
G.add_node("Chain of Thought (CoT) Prompting", size=1000)
G.add_node("Bootstrapping Prompting", size=1000)
G.add_node("RAG (Retrieval-Augmented Generation)", size=1000)
G.add_node("ReAct (Reasoning and \nActing)", size=1000)

# Add edges to represent hierarchy
G.add_edges_from([
    ("Prompting Techniques", "Basic Prompting Methods"),
    ("Prompting Techniques", "Advanced Prompting Techniques"),
    ("Prompting Techniques", "Interactive & Reasoning-Driven Methods"),
    ("Basic Prompting Methods", "One-Shot Prompting"),
    ("Basic Prompting Methods", "Few-Shot Prompting"),
    ("Advanced Prompting Techniques", "Chain of Thought (CoT) Prompting"),
    ("Advanced Prompting Techniques", "Bootstrapping Prompting"),
    ("Advanced Prompting Techniques", "RAG (Retrieval-Augmented Generation)"),
    ("Interactive & Reasoning-Driven Methods", "ReAct (Reasoning and \nActing)")
])

# Define custom node colors based on categories
node_colors = {
    "Prompting Techniques": "#2E8B57",
    "Basic Prompting Methods": "#4682B4",
    "Advanced Prompting Techniques": "#6A5ACD",
    "Interactive & Reasoning-Driven Methods": "#DAA520",
    "One-Shot Prompting": "#87CEFA",
    "Few-Shot Prompting": "#87CEFA",
    "Chain of Thought (CoT) Prompting": "#9370DB",
    "Bootstrapping Prompting": "#9370DB",
    "RAG (Retrieval-Augmented Generation)": "#9370DB",
    "ReAct (Reasoning and \nActing)": "#F4A460"
}

# Define node shapes based on their hierarchy level
node_shapes = {
    "Prompting Techniques": "o",
    "Basic Prompting Methods": "s",
    "Advanced Prompting Techniques": "s",
    "Interactive & Reasoning-Driven Methods": "s",
    "One-Shot Prompting": "o",
    "Few-Shot Prompting": "o",
    "Chain of Thought (CoT) Prompting": "o",
    "Bootstrapping Prompting": "o",
    "RAG (Retrieval-Augmented Generation)": "o",
    "ReAct (Reasoning and \nActing)": "o"
}

# Adjust positions of the nodes to prevent overlap and provide a clear layout
pos_adjusted = {
    "Prompting Techniques": (0, 0),
    "Basic Prompting Methods": (-2, -1),
    "Advanced Prompting Techniques": (0, -1),
    "Interactive & Reasoning-Driven Methods": (2, -1),
    "One-Shot Prompting": (-2.5, -2.2),
    "Few-Shot Prompting": (-1.5, -1.8),
    "Chain of Thought (CoT) Prompting": (-0.5, -2.2),
    "Bootstrapping Prompting": (0.5, -1.8),
    "RAG (Retrieval-Augmented Generation)": (1.5, -2.2),
    "ReAct (Reasoning and \nActing)": (2.5, -1.8),
}

# Final adjustment: remove title, decrease node size
final_node_size = 200  # Smaller than before

plt.figure(figsize=(12, 8))
ax = plt.gca()

# Draw the even smaller nodes with adjusted positions
for node in G.nodes:
    nx.draw_networkx_nodes(
        G, pos_adjusted, nodelist=[node], node_size=final_node_size,
        node_color=node_colors[node], node_shape=node_shapes[node],
        edgecolors="black", linewidths=1.5
    )

# # Draw edges
nx.draw_networkx_edges(G, pos_adjusted, arrows=False, style="dashed", alpha=0.5)

# Draw labels with offset to avoid overlap
label_pos_adjusted = {key: (x, y-0.1) for key, (x, y) in pos_adjusted.items()}
nx.draw_networkx_labels(G, label_pos_adjusted, font_size=10, font_weight="normal")


# final adjustments
ax.set_facecolor("white")
plt.axis('off')
# save the plot
plt.savefig("prompting_techniques_hierarchy.png")
plt.show()

```
