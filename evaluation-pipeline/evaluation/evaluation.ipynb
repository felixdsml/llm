## Preprocessing

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
from adjustText import adjust_text
import os


# Load the Excel file
file_path = "log_evaluations.xlsx"  # Replace with your actual file path
df = pd.read_excel("log_evaluations.xlsx")
model_sizes = pd.read_excel("model_sizes.xlsx")
# inner join the two dataframes
df = pd.merge(df, model_sizes, on="Model", how="inner")
```

```python
# all values in Model column to lowercase
df['Model'] = df['Model'].str.lower()
# Define the renaming function
def rename_model(model_name):
    model_name = re.sub(r'^llama-3-8b', 'llama3:8b', model_name)
    model_name = re.sub(r'^llama3:8b-bnb', 'llama3:8b-text-bnb', model_name)
    model_name = re.sub(r'^phi-3', 'phi3:14b', model_name)
    return model_name

# Apply the renaming function to the 'Model' column
df['Model'] = df['Model'].apply(rename_model)
```

```python
# replace 'llama3:8b-text-bnb-4bit-synthetic_text_to_sql-lora-3epochs-q5_k_m:latest' with 'llama3:8b-text-lora-q5_k_m'
df['Model'] = df['Model'].replace('llama3:8b-text-bnb-4bit-synthetic_text_to_sql-lora-3epochs-q5_k_m:latest', 'llama3:8b-text-lora-q5_k_m')
# replace 'llama3:8b-instruct-bnb-4bit-synthetic_text_to_sql-lora-3epochs-q5_k_m:latest' with 'llama3:8b-instruct-lora-q5_k_m'
df['Model'] = df['Model'].replace('llama3:8b-instruct-bnb-4bit-synthetic_text_to_sql-lora-3epochs-q5_k_m:latest', 'llama3:8b-instruct-lora-q5_k_m')
# replace 'phi3:14b-medium-4k-instruct-synthetic_text_to_sql-lora-3epochs-q5_k_m:latest' with 'phi3:14b-medium-4k-instruct-lora-q5_k_m'
df['Model'] = df['Model'].replace('phi3:14b-medium-4k-instruct-synthetic_text_to_sql-lora-3epochs-q5_k_m:latest', 'phi3:14b-medium-4k-instruct-lora-q5_k_m')
```

### Calculating the Combined Scores

```python
# Calculate combined scores and round to two decimal places
df['Test Combined Scores'] = df[['Test Correctness Scores', 'Test Executable Scores']].mean(axis=1)
df['Test Combined Scores - LabeledFewShot'] = df[['Test Correctness Scores - LabeledFewShot', 'Test Executable Scores - LabeledFewShot']].mean(axis=1)

# reduce all values to two decimal places
df = df.round(2)
```

```python
# Select the first 5 columns
first_five_columns = df.iloc[:, :5]

# Select columns with "Scores" and "Test" in their name after the first 5 columns
scores_test_columns = df.loc[:, [col for col in df.columns[5:] if 'Scores' in col and 'Test' in col and not 'BootstrapFewShot' in col ]]

# Combine the first 5 columns with the selected "Scores" columns
result_df = pd.concat([first_five_columns, scores_test_columns], axis=1)
```

## Export tables to Latexformat

```python
# write little function to str. replace (' Scores', ''), (' - LabeledFewShot', ' LFS'), ('Correctness', 'Corr.'), ('Executable', 'Exec.') for a df
def replace_columns(df):
    df.columns = df.columns.str.replace(' Scores', '')
    df.columns = df.columns.str.replace(' - LabeledFewShot', ' LFS')
    df.columns = df.columns.str.replace('Correctness', 'Corr.')
    df.columns = df.columns.str.replace('Executable', 'Exec.')
    df.columns = df.columns.str.replace('Combined', 'Comb.')
    return df
```

```python
# Function to clean the title and save the plot
def save_plot(fig, title, folder="figures"):
    # Clean the title to create a safe filename
    clean_title = re.sub(r'[^a-zA-Z0-9]', '_', title)
    plot_filename = os.path.join(folder, f"{clean_title}.png")

    # Ensure the "figures" subfolder exists
    if not os.path.exists(folder):
        os.makedirs(folder)

    # Save the plot
    fig.savefig(plot_filename, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a custom palette
custom_palette = sns.color_palette("Paired", 10, desat=0.9)

# Create a mapping from Score Type to a color
color_mapping = {
    "Test Match Scores": custom_palette[0],
    "Test Match Scores - LabeledFewShot": custom_palette[1],
    "Test Correctness Scores": custom_palette[2],
    "Test Correctness Scores - LabeledFewShot": custom_palette[3],
    "Test Executable Scores": custom_palette[4],
    "Test Executable Scores - LabeledFewShot": custom_palette[5],
    "Test Combined Scores": custom_palette[8],
    "Test Combined Scores - LabeledFewShot": custom_palette[9],
}

# Function to create grouped bar plot for scores with consistent colors
def plot_grouped_scores_consistent_color(df, score_columns, title):
    # Melt the dataframe for plotting
    melted_df = df.melt(id_vars=['Model'], value_vars=score_columns, var_name='Score Type', value_name='Score')

    # Create the plot
    plt.figure(figsize=(18, 6))
    ax = sns.barplot(x='Model', y='Score', hue='Score Type', data=melted_df, palette=[color_mapping[score] for score in score_columns])
    plt.title(title)
    plt.xticks(rotation=45)
    
    # Remove the x-axis title
    ax.set_xlabel('')

    # Adjust the bottom margin to make space for the legend
    plt.subplots_adjust(bottom=0.15)

    # Move the legend below the plot
    plt.legend(bbox_to_anchor=(0.5, -0.35), loc='upper center', ncol=3)

    # Save the plot using the save_plot function
    save_plot(plt.gcf(), title)

    # Show the plot
    plt.show()

# Plot grouped test scores
plot_grouped_scores_consistent_color(df, [
    "Test Match Scores", "Test Match Scores - LabeledFewShot", 
    "Test Correctness Scores", "Test Correctness Scores - LabeledFewShot",
    "Test Executable Scores", "Test Executable Scores - LabeledFewShot", 
], 'Test Scores by Model')

# Plot grouped correctness scores
plot_grouped_scores_consistent_color(df, [
    "Test Correctness Scores", "Test Correctness Scores - LabeledFewShot", 
], 'Correctness Test Scores by Model')

# Plot grouped match scores
plot_grouped_scores_consistent_color(df, [
    "Test Match Scores", "Test Match Scores - LabeledFewShot", 
], 'Match Test Scores by Model')

# Plot grouped combined scores
plot_grouped_scores_consistent_color(df, [
    "Test Combined Scores", "Test Combined Scores - LabeledFewShot", 
], 'Combined Test Scores by Model')

# Plot grouped exec scores
plot_grouped_scores_consistent_color(df, [
    "Test Executable Scores", "Test Executable Scores - LabeledFewShot", 
], 'Executable Test Scores by Model')
```

```python
# create a function for the times
def plot_grouped_times(df, title):
    # Melt the dataframe for plotting
    melted_df = df.melt(id_vars=['Model'], value_vars=['Total Time'], var_name='Time Type', value_name='Time')

    # Create the plot
    plt.figure(figsize=(18, 6))
    ax = sns.barplot(x='Model', y='Time', hue='Time Type', data=melted_df, palette='viridis')
    plt.title(title)
    plt.xticks(rotation=45)
    
    # Remove the x-axis title
    ax.set_xlabel('')

    # Adjust the bottom margin to make space for the legend
    plt.subplots_adjust(bottom=0.15)

    # Move the legend below the plot
    plt.legend(bbox_to_anchor=(0.5, -0.35), loc='upper center', ncol=3)

    # Save the plot using the save_plot function
    save_plot(plt.gcf(), title)

    # Show the plot
    plt.show()
    
# Plot grouped times 
plot_grouped_times(df, 'Total Time by Run')
```

```python
# Function to create grouped bar plot for scores
def plot_grouped_scores(df, score_columns, title):
    melted_df = df.melt(id_vars=['Model'], value_vars=score_columns, var_name='Score Type', value_name='Score')
    plt.figure(figsize=(18, 6))
    ax = sns.barplot(x='Model', y='Score', hue='Score Type', data=melted_df, palette='Set2')
    plt.title(title)
    plt.xticks(rotation=45)
    
    # Remove the x-axis title
    ax.set_xlabel('')

    # Adjust the bottom margin to make space for the legend
    plt.subplots_adjust(bottom=0.15)

    # Move the legend below the plot
    plt.legend(bbox_to_anchor=(0.5, -0.35), loc='upper center', ncol=3)

    # Save the plot using the save_plot function
    save_plot(plt.gcf(), title)

    # Show the plot
    plt.show()
    plt.show()
```

```python
# Calculate the maximum value for each score type
max_test_scores = df[['Test Combined Scores', 'Test Combined Scores - LabeledFewShot']].max(axis=0)

# Normalize the scores by dividing by the maximum value
df['Min-Max Normalized Test Combined Scores'] = df['Test Combined Scores'] / max_test_scores['Test Combined Scores']
df['Min-Max Normalized Test Combined Scores - LabeledFewShot'] = df['Test Combined Scores - LabeledFewShot'] / max_test_scores['Test Combined Scores - LabeledFewShot']

# Plot min-max normalized test scores
plot_grouped_scores(df, ['Min-Max Normalized Test Combined Scores', 'Min-Max Normalized Test Combined Scores - LabeledFewShot'], 'Min-Max Normalized Test Combined Scores by Model')


```

```python
# Calculate the mean and standard deviation for each score type
mean_test_scores = df[['Test Combined Scores', 'Test Combined Scores - LabeledFewShot', ]].mean(axis=0)
std_test_scores = df[['Test Combined Scores', 'Test Combined Scores - LabeledFewShot', ]].std(axis=0)

# Normalize the scores using z-score
df['Z-Score Normalized Test Combined Scores'] = (df['Test Combined Scores'] - mean_test_scores['Test Combined Scores']) / std_test_scores['Test Combined Scores']
df['Z-Score Normalized Test Combined Scores - LabeledFewShot'] = (df['Test Combined Scores - LabeledFewShot'] - mean_test_scores['Test Combined Scores - LabeledFewShot']) / std_test_scores['Test Combined Scores - LabeledFewShot']


# Plot z-score normalized test scores
plot_grouped_scores(df, ['Z-Score Normalized Test Combined Scores', 'Z-Score Normalized Test Combined Scores - LabeledFewShot', ], 'Z-Score Normalized Test Combined Scores by Model')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a custom palette
custom_palette = sns.color_palette("Paired", 10, desat=0.9)

# Create a mapping from Score Type to a color
color_mapping = {
    "Test Match Scores": custom_palette[0],
    "Test Match Scores - LabeledFewShot": custom_palette[1],
    "Test Correctness Scores": custom_palette[2],
    "Test Correctness Scores - LabeledFewShot": custom_palette[3],
    "Test Executable Scores": custom_palette[4],
    "Test Executable Scores - LabeledFewShot": custom_palette[5],
    "Test Combined Scores": custom_palette[8],
    "Test Combined Scores - LabeledFewShot": custom_palette[9],
}

def prepare_data_for_lora_plot(df):
    # Add a 'LoRA' column: 1 if the model is LoRA, 0 otherwise
    df['LoRA'] = df['Model'].apply(lambda x: 1 if 'lora' in x else 0)
    
    # Create a 'Base Model' column by removing the LoRA part
    df['Base Model'] = df['Model'].str.replace(r'lora-', '', regex=True)
    
    return df

def plot_lora_vs_nonlora_with_color(df, score_column, title):
    # Prepare the data
    df = prepare_data_for_lora_plot(df)
    
    # Use the color mapping for the score type
    palette = {0: color_mapping[score_column], 1: color_mapping[score_column + ' - LabeledFewShot']}

    # Create the plot
    plt.figure(figsize=(12, 6))
    ax = sns.barplot(x='Base Model', y=score_column, hue='LoRA', data=df, palette=palette)
    plt.title(title)
    plt.xticks(rotation=45)
    
    # Remove the x-axis title
    ax.set_xlabel('')
    
    # Adjust the bottom margin to make space for the legend
    plt.subplots_adjust(bottom=0.15)
    
    # save the plot
    save_plot(plt.gcf(), title)
    
    # Show the plot
    plt.show()
```

```python
# Prepare data and plot correctness scores, comparing LoRA and non-LoRA models with color coding
plot_lora_vs_nonlora_with_color(df, "Test Match Scores", "Match Test Scores by Model - Lora Models - paired")

plot_lora_vs_nonlora_with_color(df, "Test Correctness Scores", "Correctness Test Scores by Model - Lora Models - paired")

plot_lora_vs_nonlora_with_color(df, "Test Executable Scores", "Executable Test Scores by Model - Lora Models - paired")

plot_lora_vs_nonlora_with_color(df, "Test Combined Scores", "Combined Test Scores by Model - Lora Models - paired")


```
